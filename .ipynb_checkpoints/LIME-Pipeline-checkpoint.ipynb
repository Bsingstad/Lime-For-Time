```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import accuracy_score as acc
import numpy as np
from lime import explanation
from lime import lime_base
from lime_timeseries import LimeTimeSeriesExplanation
import math
```

## Load Coffee Dataset

```python
coffee_train = pd.read_csv('coffee_train.csv', sep=',', header=None).astype(float)
coffee_train_y = coffee_train.loc[:, 0]
coffee_train_x = coffee_train.loc[:, 1:]
coffee_test = pd.read_csv('coffee_test.csv', sep=',', header=None).astype(float)
coffee_test_y = coffee_test.loc[:, 0]
coffee_test_x = coffee_test.loc[:, 1:]
```

## Load Earthquake Dataset

```python
earthquake_train = pd.read_csv('earthquake_train.csv', sep=',', header=None).astype(float)
earthquake_train_y = earthquake_train.loc[:, 0]
earthquake_train_x = earthquake_train.loc[:, 1:]
earthquake_test = pd.read_csv('earthquake_test.csv', sep=',', header=None).astype(float)
earthquake_test_y = earthquake_test.loc[:, 0]
earthquake_test_x = earthquake_test.loc[:, 1:]
```

## Datasets

```python
for idx in range(0, len(coffee_train)):
    if coffee_train_y[idx] == 0:
        plt.plot(coffee_train_x.loc[idx, :], 'b')
    else:
        plt.plot(coffee_train_x.loc[idx, :], 'r')
plt.title('Coffee Dataset')
plt.show()

for idx in range(0, len(earthquake_train)):
    if earthquake_train_y[idx] == 0:
        plt.plot(earthquake_train_x.loc[idx, :], 'b')
    else:
        plt.plot(earthquake_train_x.loc[idx, :], 'r')
plt.title('Earthquake Dataset')
plt.show()
```

### Compute Accuracy

```python
knn = KNN()
knn.fit(coffee_train_x, coffee_train_y)
knn2 = KNN()
knn2.fit(earthquake_train_x, earthquake_train_y)
print('Accuracy KNN for coffee dataset: %f' % (acc(coffee_test_y, knn.predict(coffee_test_x))))
print('Accuracy KNN for earthquake dataset: %f' % (acc(earthquake_test_y, knn2.predict(earthquake_test_x))))
```

## Explaining Coffee Dataset

```python
idx = 5
num_features = 10
num_slices = 24
series = coffee_test_x.iloc[idx, :]
```

```python
explainer = LimeTimeSeriesExplanation(class_names=['0', '1'], feature_selection='auto')
exp = explainer.explain_instance(series, knn.predict_proba, num_features=num_features, num_samples=5000, num_slices=num_slices, 
                                 replacement_method='noise', training_set=coffee_train_x)
exp.as_list()
```

```python
values_per_slice = math.ceil(len(series) / num_slices)
plt.plot(series)
plt.plot(coffee_test_x.iloc[16, :], color='green')
for i in range(num_features):
    feature, _ = exp.as_list()[i]
    start = feature * values_per_slice
    end = start + values_per_slice
    plt.axvspan(start , end, color='red', alpha=0.3)
plt.show()
```
